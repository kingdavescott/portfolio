### Simple Linear Regression

Something I've gotten in the habit of doing when approaching these types of problems is to start simple. Generally there are a couple of features that are obviously going to be important when predicting the target variable. In our case, the above scatter plots, correlations (and intuition) indicate that the square footage of living space of the house is going to have some impact on its price. If equations are your thing, here's how we are going to be modeling the price of a house in this first pass:
$$ price_i = \beta_0 + \beta_1 sqft\_living_i + \epsilon_i$$
It's unlikely to give us the most accurate predictions, but it will be good to establish a baseline model to compare our more complicated models to later.

```python
# zoom in on the price ~ sqft_living scatterplot
plt.scatter(data['sqft_living'], data['price'], c='royalblue')
plt.title('sqft_living vs. price scatter')
plt.xlabel('sqft_living')
plt.ylabel('price')
```

One thing that sticks out in this plot is the distinct "funnel" shape of the data points. The variance in price increases as sqft_living increases. A common practice when modeling data of this type is to take the natural log of both variables and run your model on the transformed variables. Let's see what the data look like after being log-transformed.

```python
# show data log-transformed
plt.scatter(np.log(data['sqft_living']), np.log(data['price']), c='royalblue')
plt.title('ln(sqft_living) vs. ln(price) scatter')
plt.xlabel('ln(sqft_living)')
plt.ylabel('ln(price)')
```

The log-transformed variables are much more obviously linearly related than the originals so we'll stick with it. Given the transformation, our new model will be:

$$ \ln price_i = \beta_0 + \beta_1 \ln sqft\_living_i + \epsilon_i$$

The coefficient $\beta_1$ will now have the interpretation of the % change impact on price associated with a 1% increase in sqft_living.

Next, what we'll need to do is split our data into training and test sets. We'll build our model on the training data and then make predictions on the unseen test set to check for accuracy later.

```python
# create log price and sqft_living variables
data['log_sqft_living'] = np.log(data['sqft_living'])

# focus on variables of interest and split into training and test sets
y = np.log(data['price'])
X = data['log_sqft_living']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123)

# train linear regression model
lm = linear_model.LinearRegression()
lm.fit(X_train[:, np.newaxis], y_train)
beta_0 = lm.intercept_
beta_1 = lm.coef_

print('Beta_0: ' + str(beta_0) + ', Beta_1: ' + str(beta_1[0]))
```

Our simple model has estimated the following relationship between log price and log sqft_living:
$$ \ln price_i \approx 6.75 + 0.83 \ln sqft\_living_i$$
From these estimates we can say that a 10% increase in sqft_living is associated with a 8.3% increase in price on average. Let's check out regression line plotted against the data.

```python
# plot predictions on log scale
plt.scatter(X_train, y_train, c='royalblue')
plt.plot(X_train, lm.predict(X_train[:,np.newaxis]), c='orangered', linewidth=2)
plt.title('ln(sqft_living) vs. ln(price) scatter')
plt.xlabel('ln(sqft_living)')
plt.ylabel('ln(price)')
```

Now that we have our model, let's see how it performs on the unseen test data. There are many metrics out there for testing the accuracy of your model, but for now we'll use one called Mean Absolute Percent Error (MAPE):

$$ MAPE = \frac{100}{N}\sum_{i=1}^{N} \Bigg|\frac{\hat{y}_i - y_i}{y_i}\Bigg| $$
where $\hat{y}$ is the prediction from our model.

```python
# create function to calculate MAPE
def mape(pred, act):
    return 100 * np.mean(abs((pred - act) / act))
# compute mape using our simple linear model
# we'll also need to exponentiate our predictions to "un-log" the data


mape(np.exp(lm.predict(X_test[:, np.newaxis])), np.exp(y_test))
```

A MAPE of 32.9% isn't great, but we didn't expect great results because at this point we have a really simplified model. We're missing out on all of the other information about the house that's embedded in the other features!

### Multiple Linear Regression

Let's add some sophistication to our model by including additional features and hopefully pick up some explanatory power. It's a good idea to always start with your intuition and come up with some hypotheses before dumping all of the variables into your model blindly. Here are some variables that might be interesting for us to add:

- bedrooms: 1-33, additional bedrooms likely add value up to a point, should probably bucket them after 7+
- bathrooms: 0-8, same story with bedrooms, will model linearly since there are a lot of variations
- sqft_lot: larger lot size may be associated with higher price, but probably interacts with house size
- floors: 1-3.5: more floors, higher price?
- waterfront: 0-1, waterfront=1 should drive up price
- view: 0-4, believe this to be measure of how good the view is, 4's should be better than 0's
- condition: 1-5, quality measure mostlikely, 5's better than 1's
- grade: 1-13, slope of property? will want to bucket (actually found out this is pretty linear so don't need to)
- sqft_living, sqft_above, sqft_basement: sqft_living = sqft_above + sqft_basement, so we'll use sqft_above and sqft_basement separately in the model to allow for a differential impact of above ground space vs. basement space
- yr_built: could be a quadratic effect, brand new houses are likely priced above average and maybe much older houses are priced high for antique value, etc.
- yr_renovated: 0 or year renovated, probably a linear relationship, will want to create a dummy variable of renovated (y/n)
- zipcode: will use as a proxy for neighborhood, Mercer Island probably higher prices than SoDo
- lat/long: probably won't use
- date: could be used to model seasonality and trend in the housing market
- sqft_living15, sqft_lot15: not sure what these features are so will ignore for now

Now that we've become acquainted with the rest of our features let's do some feature engineering and data preprocessing.

```python
# create bins for bedrooms
def bedrooms_bin(row):
    if row['bedrooms'] >= 7:
        return '7+'
    else:
        return str(row['bedrooms'])
data['bedrooms_bin'] = data.apply(bedrooms_bin, axis=1)  

# take log of lot square footage
data['log_sqft_lot'] = np.log(data['sqft_lot'])

# create bins for floors
def floors_bin(row):
    if row['floors'] >= 2.5:
        return '2.5+'
    else:
        return str(row['floors'])
data['floors_bin'] = data.apply(floors_bin, axis=1)

# take log of sqft_above
data['log_sqft_above'] = np.log(data['sqft_above'])

# since some houses have no basement we can't take the natural log of sqft_basement by itself
# what we'll do is take the natural log where it != 0, create a basement indicator variable
# this should parse the effects
def log_basement(row):
    if row['sqft_basement'] == 0:
        return 0
    else:
        return np.log(row['sqft_basement'])
data['log_sqft_basement'] = data.apply(log_basement, axis=1)
data['basement'] = (data['sqft_basement'] != 0) * 1

# squared term for year built if we have a quadratic effect
data['yr_built_2'] = data['yr_built'] ** 2

# indicator variable for whether it was renovated or not
data['renovated'] = (data['yr_renovated'] != 0) * 1

# we'll now create dummy variables for our categorical features, dropping the most common to serve as the base case
bedroom_dummies = pd.get_dummies(data['bedrooms_bin'], prefix='bedrooms').drop('bedrooms_4', axis=1)
floor_dummies = pd.get_dummies(data['floors_bin'], prefix='floors').drop('floors_1.0', axis=1)
view_dummies = pd.get_dummies(data['view'], prefix='view').drop('view_0', axis=1)
condition_dummies = pd.get_dummies(data['condition'], prefix='condition').drop('condition_3', axis=1)
zip_dummies = pd.get_dummies(data['zipcode'], prefix='zipcode').drop('zipcode_98103', axis=1)

# merge together all of the features to create our model matrix, X
features = ['log_sqft_above', 'log_sqft_basement', 'basement', 'log_sqft_lot', 'waterfront', 'bathrooms', 'grade',
            'yr_built', 'yr_built_2', 'yr_renovated', 'renovated'
           ]

X = pd.concat([data[features], bedroom_dummies, floor_dummies, view_dummies, condition_dummies, zip_dummies], axis=1)
features = list(X.columns)
```

```python
# now we'll split the data into train and test splits again
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123)

# and retrain out regression model with the new variables
lm.fit(X_train, y_train)
coefficient_df = pd.DataFrame({'features': features, 'coefficient': lm.coef_})
coefficient_df.head(30)
```

Some observations:

- sqft: above ground sqft has a much larger impact than the basement or the lot
- a waterfront view adds 46% to the price of the house all else equal
- it's interesting that fewer bedrooms seem to have higher price points vs. more bedrooms
- also additional floors are associated with lower prices
- larger values of view score contribute positively to price
- as expected, better condition homes sell for more
- the zip code with the most negative relationship with price is 98002 which is Auburn, WA (sketchy)
- the zip code with the most positive relationship with price is 98039 which is Medina, WA (VERY nice houses)

Let's check to see if we've managed to improve the accuracy of the model by adding all these features.

```python
mape(np.exp(lm.predict(X_test)), np.exp(y_test))
```

13.8%, not too bad! We've significantly improved the accuracy of our regression model by including a lot more of the features of the house.
